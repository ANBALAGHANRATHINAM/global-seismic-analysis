```python
 # extracting data from url(usgs) and then  store it in one list

import requests
from datetime import date  # importing datetime

# getting earthquake data from usgs website

url = "https://earthquake.usgs.gov/fdsnws/event/1/query"
all_earthquakes = []  # list to store all earthquake records

# this is for get today's date
today = date.today()

# try to get the date exactly 5 years back
try:
    start_date_limit = today.replace(year=today.year - 5)  # 5 years back same day
except:
    # if problem with leap year (like feb 29)
    start_date_limit = date(today.year - 5, today.month, 28)

print("Collecting earthquakes from", start_date_limit, "to", today)      # for showing the total records till now

# loop for each year from start year to current year
for year in range(start_date_limit.year, today.year + 1):

    # loop for each month from 1 to 12 (n-1)
    for month in range(1, 13):

        # first day of this month
        month_start = date(year, month, 1)

        # find first day of next month
        if month == 12:
            month_end = date(year + 1, 1, 1)
        else:
            month_end = date(year, month + 1, 1)

        # skip if this month is fully before the start date
        if month_end <= start_date_limit:
            continue

        # skip if this month is after today (future)
        if month_start > today:
            continue

        # fix start date if it is before the allowed limit
        if month_start < start_date_limit:
            month_start = start_date_limit

        # fix end date if it goes beyond today
        if month_end > today:
            month_end = today

        print("Fetching:", month_start, "to", month_end)

        # parameters for the api request
        parameters = {
            "format": "geojson",
            "starttime": month_start.isoformat(),
            "endtime": month_end.isoformat(),
            "minmagnitude": 3
        }

        # make the api request (call)
        response = requests.get(url, params=parameters)

        # check if request is successful
        if response.status_code == 200:
            json_data = response.json()
            monthly_records = json_data.get("features", [])
            print("Events found:", len(monthly_records))

            # adding  all monthly records to main list
            all_earthquakes.extend(monthly_records)
        else:
            print("Request failed with status:", response.status_code)

# final count of all earthquake records
print("\nTotal earthquake records collected:", len(all_earthquakes))
```

Data cleaning and Handling 

Name changing 

```python
df.columns=['Type', 'id', 'mag', 'place', 'time',                                     # changing column names 
       'updated', 'tz', 'url',
       'detail', 'felt', 'cdi',
       'mmi', 'alert', 'status',
       'tsunami', 'sig', 'net',
       'code', 'ids', 'sources',
       'types', 'nst', 'dmin',
       'rms', 'gap', 'magType',
       'e_type', 'title', 'gtype',
       'coordinates']
```

```python
df.columns
```

deriving new columns 

```python
df["time"] = pd.to_datetime(df["time"], unit='ms')        
df["updated"] = pd.to_datetime(df["updated"], unit='ms')                              # converting time from milliseconds to datetime
```

```python
df[['time','updated']]
```

```python
df["longitude"] = df["coordinates"].apply(lambda x: x[0])
df["latitude"]  = df["coordinates"].apply(lambda x: x[1])                             # extracting values and deriving new columns 
df["depth_km"]  = df["coordinates"].apply(lambda x: x[2])
```

```python
df[["longitude","latitude","depth_km"]]
```

```python
# to seperate country name
df['country'] = df['place'].str.split(',').str[-1].str.strip()
```

```python
df["country"]
```

```python
df['year'] = df['time'].dt.year                                                       # separating year from time
```

```python
df['month'] = df['time'].dt.month                                                     # separating month from time
```

```python
df['day'] = df['time'].dt.day                                                         # separating day from time
```

```python
df['day_of_week'] = df['time'].dt.day_name()                                          # separating day_of_week from time
```

```python
df[['day_of_week','year','month','day']]
```

```python
df['depth_flag'] = df['depth_km'].apply(lambda x: 'shallow' if x <= 70 else 'deep')   # creating Shallow/deep earthquake flag column based on depth_km.
```

```python
df['mag_flag'] = df['mag'].apply(lambda x: 'destructive' if x >=6 else 'strong')      # creating Strong/destructive flag columns based on mag thresholds.
```

```python
df[['mag_flag','depth_flag']]
```

```python
df.columns
```

Data correction 

```python
df['felt'] = df['felt'].fillna(0)                      #Filling missing numeric values with 0
```

```python
df['cdi'] = df['cdi'].fillna(0)                        #Fill missing numeric values with 0
```

```python
df['mmi'] = df['mmi'].fillna(0)                        #Fill missing numeric values with 0
```

```python
df[["felt","cdi","mmi"]].isna().sum()
```

```python
df['dmin'] = df['dmin'].fillna(df['dmin'].mean())      #replacing NaN values with mean values 
```

```python
df['nst'] = df['nst'].fillna(df['nst'].mean())         #replacing NaN values with mean values 
```

```python
df['gap'] = df['gap'].fillna(df['gap'].mean())         #replacing NaN values with mean values 
```

```python
df['rms'] = df['rms'].fillna(df['rms'].mean())         #replacing NaN values with mean values 
```

```python
df[["dmin","nst","gap","rms"]].isna().sum()
```

```python
df.drop(["Type","tz","url","detail","coordinates","title","gtype"],axis=1,inplace=True) #  dropping columns 
```

```python
arranged_column = ["id","time","updated","year","month","day","day_of_week","mag","mag_flag","magType","depth_km","depth_flag","longitude","latitude","place","country","e_type","tsunami","sources","net","alert","felt","cdi","mmi","types","sig","code","ids","nst","dmin","rms","gap","status"]
df = df[arranged_column]                               # arranging columns for better understanding 
```
